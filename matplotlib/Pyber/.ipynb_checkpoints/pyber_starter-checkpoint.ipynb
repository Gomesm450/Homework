```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

df_city = pd.read_csv(city_data)
df_ride = pd.read_csv(ride_data)
df_ride.head()
df_city.head()


df_merge = pd.merge(df_ride, df_city, on="city")
df_merge.head()
```

## Bubble Plot of Ride Sharing Data

```python

urban_df = df_merge[df_merge["type"] == "Urban"]
city_urban_df = urban_df.groupby(["city"])
x_axis_U = city_urban_df["ride_id"].count()
y_axis_U = city_urban_df["fare"].mean()
z_U = city_urban_df["driver_count"].sum()


suburb_df = df_merge[df_merge["type"] == "Suburban"]
city_suburb_df = suburb_df.groupby(["city"])
x_axis_S = city_suburb_df["ride_id"].count()
y_axis_S = city_suburb_df["fare"].mean()
z_S = city_suburb_df["driver_count"].sum()


rural_df = df_merge[df_merge["type"] == "Rural"]
city_rural_df = rural_df.groupby(["city"])
x_axis_R = city_rural_df["ride_id"].count()
y_axis_R = city_rural_df["fare"].mean()
z_R = city_rural_df["driver_count"].sum()


plt.scatter(x_axis_U, y_axis_U, s=z_U, c="gold",edgecolors="black", alpha=0.6, linewidth=1, label = "Urban")
plt.scatter(x_axis_S, y_axis_S, s=z_S, c="skyblue",edgecolors="black", alpha=0.6, linewidth=1, label = "Suburban")
plt.scatter(x_axis_R, y_axis_R, s=z_R, c="lightcoral",edgecolors="black", alpha=0.6, linewidth=1, label = "Rural")


plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

ld = plt.legend(loc="upper right", scatterpoints=1, fontsize=10, title = "City Types")
ld.legendHandles[0]._sizes = [30]
ld.legendHandles[1]._sizes = [30]
ld.legendHandles[2]._sizes = [30]


plt.savefig("data/bubble.png", bbox_inches="tight")
```

## Total Fares by City Type

```python

fare_U = urban_df["fare"].sum()
fare_S = suburb_df["fare"].sum()
fare_R = rural_df["fare"].sum()


label = ["Urban", "Suburban", "Rural"]
size = [fare_U, fare_S, fare_R]
color = ["gold", "lightskyblue", "lightcoral"]
plt.title('% of Total Fares by City Type')
plt.pie(sizes, labels=label, colors=color,
        autopct="%1.1f%%", shadow=True, startangle=270)


plt.savefig("data/pie.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
fare_U = urban_df["fare"].count()
fare_S = suburb_df["fare"].count()
fare_R = rural_df["fare"].count()


# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [fare_U, fare_S, fare_R]
colors = ["gold", "lightskyblue", "lightcoral"]
plt.title('% of Total Rides by City Type')
plt.pie(sizes,labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
# Save Figure
plt.savefig("data/pie2.png")
```

## Total Drivers by City Type
